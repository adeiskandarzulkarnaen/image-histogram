##  reading images
<!-- Membaca gambar -->

```python
img = cv2.imread("ana-fi-bandung.jpg")
print('Image Resolution : ', img.shape)
```

## showing image
<!-- Menampilkan Gambar -->

```python
cv2_imshow(img)
```

## make histogram plot
<!-- membuat plot histogram -->

```python
plt.hist(img.ravel(),256,[0,256]); 
plt.show()
```

```python
# rgb histogram
color = ('r','g','b')
for i,col in enumerate(color):
  histr = cv2.calcHist([img],[i],None,[256],[0,256])
  plt.plot(histr,color = col)
  plt.xlim([0,256])
plt.show()
```

## Conclusion
Based on the histogram plot, it can be concluded that my photo belongs to the **under exposed** image .
